## Question 1: What is hypothesis testing in statistics?

Hypothesis testing is a formal statistical procedure used to make decisions about populations based on sample data. It involves formulating competing hypotheses about a population parameter and using sample statistics to determine which hypothesis is more likely to be true.

Key components of hypothesis testing:

1. **Null hypothesis (H₀)**: A specific statement about a population parameter that represents the status quo or no effect/difference.

2. **Alternative hypothesis (H₁ or Hₐ)**: A statement that contradicts the null hypothesis and represents what the researcher suspects might be true.

3. **Test statistic**: A value calculated from sample data used to determine whether to reject the null hypothesis.

4. **Significance level (α)**: The threshold probability for rejecting the null hypothesis when it is actually true.

5. **P-value**: The probability of obtaining the observed results (or more extreme) if the null hypothesis were true.

6. **Decision rule**: Reject H₀ if p-value < α; otherwise, fail to reject H₀.

Common applications include testing for differences between groups, testing for relationships between variables, and testing for specific parameter values.

## Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Null Hypothesis (H₀)**:
- The statement being tested in hypothesis testing
- Represents the status quo, no effect, no difference, or no relationship
- Always contains an equality (=, ≤, or ≥)
- Assumed to be true until evidence suggests otherwise
- The hypothesis we try to find evidence against

**Alternative Hypothesis (H₁ or Hₐ)**:
- The statement that contradicts the null hypothesis
- Represents the research question or what the researcher suspects is true
- Always contains an inequality (≠, >, or <)
- Accepted only when there is sufficient evidence to reject the null hypothesis
- The hypothesis we are gathering evidence to support

**Key Differences**:

1. **Purpose**:
   - H₀: Used as a baseline for comparison
   - H₁: Represents the research question or suspected effect

2. **Statistical Support**:
   - H₀: Can only fail to be rejected or be rejected, never "proven" or "accepted"
   - H₁: Supported indirectly through rejection of H₀

3. **Formulation**:
   - H₀: Always includes equality (=, ≤, ≥)
   - H₁: Always includes inequality (≠, >, <)

4. **Default Position**:
   - H₀: Assumed true until proven otherwise (presumption of innocence)
   - H₁: Requires evidence to support

## Question 3: What is the significance level in hypothesis testing, and why is it important?

The significance level (denoted by α, alpha) is the threshold probability for rejecting the null hypothesis when it is actually true. It represents the maximum acceptable probability of making a Type I error (false positive).

**Definition and Representation**:
- Typically set at 0.05 (5%), 0.01 (1%), or 0.10 (10%)
- Determines the critical region in the sampling distribution
- Sets the threshold for the p-value below which the null hypothesis is rejected

**Importance of the Significance Level**:

1. **Error Control**: Directly controls the probability of Type I error (rejecting a true null hypothesis)

2. **Decision Framework**: Provides a clear, objective criterion for hypothesis test decisions

3. **Scientific Rigor**: Establishes standards for evidence needed to claim a significant finding

4. **Balancing Errors**: Helps balance the tradeoff between Type I and Type II errors based on consequences

5. **Reproducibility**: Promotes consistency in research by standardizing the decision threshold

6. **Confidence Level Relationship**: Directly related to confidence level (1-α) in interval estimation

The choice of significance level should reflect the relative costs of Type I versus Type II errors in the specific context of the research or decision-making scenario.

## Question 4: What does a P-value represent in hypothesis testing?

A p-value is the probability of obtaining the observed results, or more extreme results, if the null hypothesis were true. It quantifies the strength of evidence against the null hypothesis.

**Formal Definition**:
- The p-value is the probability, under the assumption that the null hypothesis is true, of obtaining a test statistic at least as extreme as the one actually observed.

**Key Characteristics**:

1. **Conditional Probability**: It is calculated assuming the null hypothesis is true.

2. **Measure of Evidence**: Smaller p-values indicate stronger evidence against the null hypothesis.

3. **Based on Observed Data**: Calculated from the actual sample data and test statistic.

4. **Considers More Extreme Results**: Includes the probability of the observed result plus more extreme outcomes.

5. **Range**: Always between 0 and 1 (since it's a probability).

6. **Decision Criterion**: Compared to the pre-specified significance level (α) to make a decision:
   - If p-value ≤ α: Reject the null hypothesis
   - If p-value > α: Fail to reject the null hypothesis

The p-value does NOT represent the probability that the null hypothesis is true, nor does it represent the probability that the alternative hypothesis is false.

## Question 5: How do you interpret the P-value in hypothesis testing?

Proper interpretation of p-values is crucial for sound statistical inference. Here's how to interpret p-values correctly:

**Correct Interpretations**:

1. **Evidence Strength**: The p-value measures the strength of evidence against the null hypothesis - smaller p-values indicate stronger evidence.

2. **Probability Statement**: It is the probability of obtaining the observed results or more extreme results if the null hypothesis were true.

3. **Relative to Significance Level**:
   - p-value ≤ α: Results are statistically significant; reject null hypothesis
   - p-value > α: Results are not statistically significant; fail to reject null hypothesis

4. **Continuous Measure**: The p-value can be viewed as a continuous measure of evidence rather than just a binary decision tool.

**Common Misinterpretations to Avoid**:

1. The p-value is NOT the probability that the null hypothesis is true.

2. The p-value is NOT the probability that the alternative hypothesis is false.

3. The p-value is NOT the probability of obtaining the same results if the experiment were repeated.

4. 1-p is NOT the probability that the alternative hypothesis is true.

5. A small p-value does NOT indicate a large or important effect size.

6. A large p-value does NOT prove the null hypothesis; it only fails to reject it.

**Contextual Considerations**:
- Statistical significance (low p-value) does not necessarily imply practical significance
- P-values should be considered alongside effect size, confidence intervals, and domain knowledge
- Multiple testing increases the chance of false positives, requiring adjustments

## Question 6: What are Type 1 and Type 2 errors in hypothesis testing?

Type I and Type II errors are the two possible mistakes that can occur in hypothesis testing.

**Type I Error (False Positive)**:
- Definition: Rejecting a null hypothesis when it is actually true
- Probability: Denoted by α (alpha), the significance level
- Example: Concluding a new drug is effective when it actually isn't
- Analogy: Convicting an innocent person in court

**Type II Error (False Negative)**:
- Definition: Failing to reject a null hypothesis when it is actually false
- Probability: Denoted by β (beta)
- Example: Concluding a new drug is ineffective when it actually works
- Analogy: Acquitting a guilty person in court

**Related Concept: Power = 1-β**:
- Power is the probability of correctly rejecting a false null hypothesis
- Higher power means lower chance of Type II error

**Relationship Between Error Types**:
- Trade-off: Decreasing the probability of one type of error usually increases the probability of the other
- For a fixed sample size, decreasing α (reducing Type I error risk) increases β (increasing Type II error risk)
- Increasing sample size can reduce both types of errors simultaneously

**Controlling Errors**:
- Type I error is controlled directly by setting the significance level (α)
- Type II error is controlled indirectly through sample size, effect size, and chosen α

## Question 7: What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

One-tailed and two-tailed tests differ in how they define the critical region and what types of alternative hypotheses they test.

**Two-Tailed Test**:
- Alternative hypothesis (H₁): Parameter ≠ specified value
- Critical region: Split into both tails of the distribution (e.g., lowest 2.5% and highest 2.5% for α = 0.05)
- Detects differences in either direction (above or below the null value)
- P-value: Probability of obtaining a test statistic as extreme or more extreme in either direction
- Example H₀: μ = 100; H₁: μ ≠ 100

**One-Tailed Test**:
- Alternative hypothesis (H₁): Parameter > specified value (right-tailed) OR Parameter < specified value (left-tailed)
- Critical region: Entirely in one tail of the distribution (e.g., highest 5% for right-tailed test with α = 0.05)
- Detects differences in only one direction (only above OR only below the null value)
- P-value: Probability of obtaining a test statistic as extreme or more extreme in the specified direction
- Example right-tailed H₀: μ ≤ 100; H₁: μ > 100
- Example left-tailed H₀: μ ≥ 100; H₁: μ < 100

**Key Differences**:

1. **Critical Region Placement**:
   - Two-tailed: Both tails of distribution
   - One-tailed: Single tail of distribution

2. **Power Differences**:
   - One-tailed: More powerful when the direction of effect is correctly anticipated
   - Two-tailed: Less powerful but protects against effects in either direction

3. **Usage Guidelines**:
   - Two-tailed: Used when there's no prior expectation about direction of effect
   - One-tailed: Used when theory or previous research strongly suggests direction of effect

## Question 8: What is the Z-test, and when is it used in hypothesis testing?

A Z-test is a statistical procedure used to test hypotheses about population parameters when the population standard deviation is known or when the sample size is large enough (typically n ≥ 30).

**Definition and Calculation**:
- Z-statistic = (Sample statistic - Hypothesized parameter value) / Standard error
- For means: Z = (x̄ - μ₀) / (σ/√n), where x̄ is sample mean, μ₀ is hypothesized population mean, σ is population standard deviation, and n is sample size
- For proportions: Z = (p̂ - p₀) / √[p₀(1-p₀)/n], where p̂ is sample proportion, p₀ is hypothesized population proportion, and n is sample size

**When to Use Z-tests**:

1. **Known Population Standard Deviation**: When σ is known (rare in practice)

2. **Large Sample Size**: When n ≥ 30, so the Central Limit Theorem applies and the sampling distribution of the mean is approximately normal

3. **Testing Means**: One-sample, two-sample, or paired sample tests for population means

4. **Testing Proportions**: Tests involving population proportions (regardless of sample size, provided np ≥ 5 and n(1-p) ≥ 5)

5. **Normal Distribution**: When the population distribution is known to be normal (for smaller samples)

**Common Z-test Applications**:
- One-sample Z-test for means: Testing if a sample mean differs from a population mean
- Two-sample Z-test for means: Testing if two population means differ
- Z-test for proportions: Testing if a sample proportion differs from a hypothesized value
- Z-test for comparing two proportions: Testing if two population proportions differ

## Question 9: How do you calculate the Z-score, and what does it represent in hypothesis testing?

**Z-score Calculation**:

The Z-score (also called Z-statistic) is calculated as:

Z = (Sample statistic - Hypothesized parameter value) / Standard error

Specific formulas:
- For means: Z = (x̄ - μ₀) / (σ/√n)
  - x̄ = sample mean
  - μ₀ = hypothesized population mean
  - σ = population standard deviation
  - n = sample size

- For proportions: Z = (p̂ - p₀) / √[p₀(1-p₀)/n]
  - p̂ = sample proportion
  - p₀ = hypothesized population proportion
  - n = sample size

**What Z-scores Represent in Hypothesis Testing**:

1. **Standardized Distance**: The Z-score represents how many standard errors the sample statistic is from the hypothesized parameter value.

2. **Evidence Quantification**: Larger absolute Z-values indicate stronger evidence against the null hypothesis.

3. **P-value Determinant**: The Z-score is used to find the p-value by consulting the standard normal distribution table or function.

4. **Decision Criterion**: The Z-score is compared to critical Z-values to determine whether to reject the null hypothesis:
   - For two-tailed tests: Reject H₀ if |Z| > Z_{α/2}
   - For right-tailed tests: Reject H₀ if Z > Z_{α}
   - For left-tailed tests: Reject H₀ if Z < -Z_{α}

5. **Effect Direction**: The sign of the Z-score indicates the direction of the difference between the sample statistic and the hypothesized value.

6. **Comparability**: Z-scores allow for standardized comparison across different hypothesis tests.

## Question 10: What is the T-distribution, and when should it be used instead of the normal distribution?

The t-distribution (Student's t-distribution) is a probability distribution that resembles the standard normal distribution but has heavier tails, meaning it has more probability in the extremes.

**Key Characteristics of the t-distribution**:
- Bell-shaped and symmetric around mean 0
- Family of distributions defined by degrees of freedom (df)
- As df increases, the t-distribution approaches the standard normal distribution
- Wider and flatter than normal distribution for small df
- When df > 30, t-distribution is very close to normal distribution

**When to Use the t-distribution Instead of Normal Distribution**:

1. **Unknown Population Standard Deviation**: When σ is unknown and must be estimated from the sample (s)

2. **Small Sample Size**: Particularly when n < 30

3. **Inference About Means**: For hypothesis tests and confidence intervals about means

4. **Regression Analysis**: For tests of significance in regression coefficients

5. **Paired Comparisons**: When performing a paired difference test

**Mathematical Relationship**:
- If Z follows a standard normal distribution and V follows a chi-squared distribution with v degrees of freedom, then Z/√(V/v) follows a t-distribution with v degrees of freedom
- This mathematical relationship explains why the t-distribution accounts for the additional uncertainty from estimating the population standard deviation

## Question 11: What is the difference between a Z-test and a T-test?

Z-tests and t-tests are both used for hypothesis testing about means, but they differ in several important ways:

**Key Differences**:

1. **Distribution**:
   - Z-test: Uses the standard normal distribution (Z-distribution)
   - T-test: Uses Student's t-distribution

2. **Population Standard Deviation Requirement**:
   - Z-test: Requires known population standard deviation (σ)
   - T-test: Uses estimated standard deviation from the sample (s)

3. **Sample Size Considerations**:
   - Z-test: Typically used when n ≥ 30 (if σ is unknown)
   - T-test: Can be used with any sample size, but especially important when n < 30

4. **Critical Values**:
   - Z-test: Critical values are fixed (e.g., ±1.96 for 95% confidence)
   - T-test: Critical values depend on degrees of freedom and decrease as df increases

5. **Degrees of Freedom**:
   - Z-test: Not applicable
   - T-test: Requires calculation of appropriate degrees of freedom

6. **Common Applications**:
   - Z-test: Proportion tests, large sample mean tests, known σ scenarios
   - T-test: Small sample mean tests, comparing group means, regression coefficients

7. **Robustness**:
   - Z-test: More powerful when assumptions are met
   - T-test: More conservative and robust when population parameters are unknown

## Question 12: What is the T-test, and how is it used in hypothesis testing?

The t-test is a statistical hypothesis test used to determine if there is a significant difference between means when the population standard deviation is unknown.

**T-statistic Formula**:
- One-sample: t = (x̄ - μ₀) / (s/√n)
  - x̄ = sample mean
  - μ₀ = hypothesized population mean
  - s = sample standard deviation
  - n = sample size

- Two-sample (independent, equal variances): t = (x̄₁ - x̄₂) / (sp√(1/n₁ + 1/n₂))
  - sp = pooled standard deviation
  - df = n₁ + n₂ - 2

- Two-sample (independent, unequal variances - Welch's t-test): t = (x̄₁ - x̄₂) / √(s₁²/n₁ + s₂²/n₂)
  - df calculated by Welch-Satterthwaite equation

- Paired: t = d̄ / (sd/√n), where d̄ is mean difference
  - df = n - 1

**Types of T-tests**:

1. **One-Sample T-test**:
   - Compares a sample mean to a known or hypothesized population mean
   - Used when the population standard deviation is unknown

2. **Independent Two-Sample T-test**:
   - Compares means from two independent groups
   - Two variations: equal variances assumed or unequal variances (Welch's t-test)

3. **Paired Sample T-test**:
   - Compares means from the same group at different times or under different conditions
   - Tests for significant mean difference between paired observations

**Applications in Hypothesis Testing**:
- Testing if a sample mean differs from a population mean
- Testing if means differ between experimental and control groups
- Testing for pre-post treatment effects
- Testing regression coefficients for significance
- Comparing product performance or customer satisfaction

## Question 14: What is the relationship between Z-test and T-test in hypothesis testing?

The Z-test and t-test are related statistical procedures with important connections and differences.

**Mathematical Relationship**:

1. **Limiting Behavior**: As sample size increases, the t-distribution approaches the standard normal (Z) distribution:
   - When df → ∞, the t-distribution → Z-distribution
   - Practically, when n > 30, t-critical values are very close to Z-critical values

2. **Formulas**:
   - Z-statistic: Z = (x̄ - μ₀) / (σ/√n)
   - t-statistic: t = (x̄ - μ₀) / (s/√n)
   - The only difference is using population standard deviation (σ) versus sample standard deviation (s)

**Practical Relationships**:

1. **Substitution Relationship**:
   - The t-test is used when we don't know σ and must estimate it with s
   - Z-test can be seen as a special case of t-test when σ is known

2. **Degree of Conservatism**:
   - t-test is more conservative than Z-test for small samples
   - t-critical values > Z-critical values for the same significance level
   - This difference accounts for the additional uncertainty when estimating σ

3. **Selection Criteria**:
   - Known σ: Use Z-test
   - Unknown σ, large sample (n ≥ 30): Either test works (practically equivalent)
   - Unknown σ, small sample (n < 30): Must use t-test

4. **Areas of Application**:
   - Z-test: Used more for proportions, large samples
   - t-test: Used more for means with unknown population parameters

## Q14   What is a confidence interval, and how is it used to interpret statistical results? 
A **confidence interval (CI)** is a range of values, derived from sample data, that is likely to contain the true population parameter (such as a mean or proportion) with a specified level of confidence.

- It gives an estimated range believed to contain the true value.
- The confidence level (commonly 95%) represents the probability that the interval calculated from repeated samples will contain the true population parameter.

---

### How is it Used to Interpret Statistical Results?

- If a 95% confidence interval for a mean is (10, 20), it means we are 95% confident that the true population mean lies between 10 and 20.
- Wider intervals indicate more uncertainty; narrower intervals indicate more precision.
- Confidence intervals help assess **statistical significance**:
  - For example, if a 95% CI for a difference between groups does **not** include zero, it suggests a statistically significant difference.
- They provide more information than just a point estimate by showing the reliability of the estimate.

---

| Term                 | Explanation                           |
|----------------------|-------------------------------------|
| Confidence Interval   | Range likely to contain true value  |
| Confidence Level     | Probability the interval is correct |

## Q 16 .How is Bayes' Theorem used in statistics, and what is its significance
# Bayes' Theorem in Statistics

Bayes' Theorem is a fundamental concept in statistics that describes how to update the probability of a hypothesis based on new evidence or data. It relates the conditional and marginal probabilities of random events and is expressed mathematically as:

\[
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
\]

Where:  
- **\(P(A|B)\)** is the *posterior probability*: the probability of event \(A\) given that event \(B\) has occurred.  
- **\(P(B|A)\)** is the *likelihood*: the probability of observing event \(B\) given that \(A\) is true.  
- **\(P(A)\)** is the *prior probability*: the initial probability of event \(A\) before seeing the evidence \(B\).  
- **\(P(B)\)** is the *marginal probability* of event \(B\).

## Use in Statistics:
- **Updating beliefs:** Bayes' theorem allows statisticians to update the probability estimate for a hypothesis as more data becomes available.
- **Decision making:** It helps in decision-making processes by incorporating prior knowledge and observed data.
- **Bayesian inference:** It forms the basis of Bayesian statistics, which contrasts with frequentist methods by treating parameters as random variables.

## Significance:
- Provides a formal mechanism for combining prior knowledge with new evidence.
- Widely used in machine learning, medical diagnosis, spam filtering, and risk assessment.
- Enables probabilistic reasoning in uncertain environments.


## 17. What is the Chi-square Distribution, and When Is It Used?

**Chi-square (χ²) distribution** is a continuous probability distribution that describes the distribution of a sum of the squares of \(k\) independent standard normal random variables.

- It is defined by its **degrees of freedom (df)**, which usually equals the number of independent variables.
- The shape is right-skewed and depends on the degrees of freedom.
- It is widely used in statistical hypothesis testing involving categorical data.

---

#### Formula:

If \(Z_1, Z_2, ..., Z_k\) are independent standard normal variables, then the Chi-square variable is:

\[
\chi^2 = \sum_{i=1}^{k} Z_i^2
\]

where \(k\) = degrees of freedom.

---

#### When Is It Used?

- **Goodness of Fit Test:** To check if observed frequencies fit an expected distribution.
- **Test of Independence:** To test the association between two categorical variables in a contingency table.
- **Test of Homogeneity:** To compare distributions of categorical variables across different populations.
- **Variance Estimation:** Used for inference about population variance in normally distributed data.

---

#### Key Properties:

- The Chi-square distribution is **only defined for positive values** (\(\chi^2 \geq 0\)).
- As degrees of freedom increase, the distribution becomes more symmetric and approaches a normal distribution.


## 18. What is the Chi-square Goodness of Fit Test, and How Is It Applied?

- The **Chi-square goodness of fit test** checks if observed data fits a specific expected distribution.
- Steps:  
  1. Define hypotheses (null: data fits expected distribution).  
  2. Calculate expected frequencies.  
  3. Compute test statistic:  
     \[
     \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
     \]  
     where \(O_i\) = observed frequency, \(E_i\) = expected frequency.  
  4. Compare with critical value from Chi-square distribution.  
- **Application:**  
  To see if sample data matches a theoretical distribution, e.g., dice rolls, genetic traits.


## 19. What is the F-distribution, and When Is It Used in Hypothesis Testing?

**F-distribution** is a continuous probability distribution that arises as the ratio of two independent chi-square distributed variables divided by their respective degrees of freedom.

---

#### Definition and Formula:

If  
- \(X_1 \sim \chi^2(d_1)\) with degrees of freedom \(d_1\),  
- \(X_2 \sim \chi^2(d_2)\) with degrees of freedom \(d_2\),  

then the **F-distributed variable** is:

\[
F = \frac{(X_1 / d_1)}{(X_2 / d_2)}
\]

where:  
- \(d_1\) = numerator degrees of freedom,  
- \(d_2\) = denominator degrees of freedom.

---

#### Characteristics:

- The F-distribution is **right-skewed** and defined only for positive values (\(F \geq 0\)).
- Shape depends on the degrees of freedom \(d_1\) and \(d_2\).
- Used to compare variances and analyze variance between groups.

---

#### When Is It Used in Hypothesis Testing?

- **ANOVA (Analysis of Variance):** To test whether the means of three or more groups are equal.
- **F-test for Equality of Variances:** To test if two populations have equal variances.
- **Regression Analysis:** To test the overall significance of a regression model.

---

#### Example of Hypothesis in ANOVA:

- Null Hypothesis \(H_0\): All group means are equal.
- Alternative Hypothesis \(H_a\): At least one group mean differs.

If the calculated \(F\)-statistic is greater than the critical value from the F-distribution table (based on degrees of freedom and significance level), \(H_0\) is rejected.

---

### Summary:

- The F-distribution helps in comparing **ratios of variances**.
- It is essential in tests involving multiple group comparisons and model significances.

## Q 20 What is an ANOVA Test, and What Are Its Assumptions?

**ANOVA (Analysis of Variance)** is a statistical method used to test whether there are any statistically significant differences between the means of three or more independent groups.

---

#### Purpose:
- To compare the means of multiple groups simultaneously.
- Helps determine if at least one group mean is different from the others.

---

#### How ANOVA Works:

- It splits the total variability in data into:
  - **Between-group variability** (differences among group means)
  - **Within-group variability** (variability within each group)

- Computes the **F-statistic** as the ratio of between-group variance to within-group variance:

\[
F = \frac{\text{Mean Square Between Groups (MSB)}}{\text{Mean Square Within Groups (MSW)}}
\]

where  
\[
\text{MSB} = \frac{\text{Sum of Squares Between Groups (SSB)}}{\text{df between}} \quad , \quad
\text{MSW} = \frac{\text{Sum of Squares Within Groups (SSW)}}{\text{df within}}
\]

- Degrees of freedom:  
  \[
  df_{\text{between}} = k - 1
  \]
  \[
  df_{\text{within}} = N - k
  \]
  where \(k\) = number of groups, and \(N\) = total observations.

---

#### Assumptions of ANOVA:

1. **Independence of Observations**  
   - Samples are independent of each other.

2. **Normality**  
   - Data in each group is approximately normally distributed.

3. **Homogeneity of Variances (Homoscedasticity)**  
   - Variances within each group are approximately equal.

---

Violating these assumptions can affect the validity of ANOVA results, and alternative methods (like non-parametric tests) might be required.


## Q21 What Are the Different Types of ANOVA Tests?

ANOVA has several types depending on the design of the experiment and the number of factors involved:

---

#### 1. One-Way ANOVA
- Used when comparing the means of **three or more independent groups** based on **one categorical independent variable** (factor).
- Example: Comparing test scores of students from three different teaching methods.

---

#### 2. Two-Way ANOVA
- Used when there are **two independent categorical variables (factors)**.
- Tests for:
  - The **main effect** of each factor on the dependent variable.
  - The **interaction effect** between the two factors.
- Example: Studying the effect of teaching method (factor 1) and gender (factor 2) on test scores.

---

#### 3. Repeated Measures ANOVA
- Used when the **same subjects are measured multiple times** under different conditions or over time.
- Controls for subject variability.
- Example: Measuring blood pressure of patients before, during, and after treatment.

---

#### 4. Mixed-Design ANOVA (Split-Plot ANOVA)
- Combines **between-subjects factors** (different groups) and **within-subjects factors** (repeated measures).
- Example: Testing effects of a drug (between-subject) over time (within-subject).

---

## Q22 What is the F-test, and How Does It Relate to Hypothesis Testing?

---

#### What is the F-test?

The **F-test** is a statistical test that uses the **F-distribution** to compare variances and test hypotheses about the equality of group means or variances.

It is commonly used to determine if:

- Two population variances are equal (in an F-test for equality of variances).
- Means across multiple groups are equal (in ANOVA).
- The overall significance of a regression model.

---

#### How It Works:

- The F-test calculates an **F-statistic**, which is a ratio of two variances:

\[
F = \frac{\text{Variance estimate 1}}{\text{Variance estimate 2}}
\]

- This ratio compares variability **between groups** to variability **within groups** (or between two variances).

- A larger F-statistic suggests greater difference between groups relative to within groups.

---

#### Relation to Hypothesis Testing:

- **Null Hypothesis \(H_0\):** The variances (or group means) are equal.
- **Alternative Hypothesis \(H_a\):** At least one variance (or group mean) is different.

- After calculating the F-statistic, it is compared to the critical value from the F-distribution table based on degrees of freedom and significance level (\(\alpha\)).

- If the F-statistic is **greater than** the critical value, **reject \(H_0\)**, indicating significant differences.

---

#### Summary:

| Purpose                  | Test Type                     | Null Hypothesis               |
|--------------------------|-------------------------------|------------------------------|
| Equality of variances    | F-test for variances          | Variances are equal           |
| Equality of group means  | ANOVA (uses F-test)           | Means are equal               |
| Overall regression model | F-test in regression analysis | Model is not significant      |

## Question 6: Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

```python
# Parameters for discrete uniform distribution
low = 1
high = 10
num_samples = 10000

# Generate random samples from discrete uniform distribution
discrete_uniform_samples = np.random.randint(low, high+1, size=num_samples)

# Function to calculate empirical CDF
def empirical_cdf(data, x):
    """
    Calculate the empirical CDF at point x.
    
    Parameters:
    data (array): Sample data
    x (float): Point at which to evaluate the CDF
    
    Returns:
    float: CDF value at x
    """
    return np.mean(data <= x)

# Calculate empirical CDF values
x_values = np.arange(low-1, high+2)  # Include points before and after the range
empirical_cdf_values = [empirical_cdf(discrete_uniform_samples, x) for x in x_values]

# Calculate theoretical CDF values
def theoretical_discrete_uniform_cdf(x, low, high):
    """
    Calculate the theoretical CDF of a discrete uniform distribution.
    
    Parameters:
    x (float): Point at which to evaluate the CDF
    low (int): Lower bound of the distribution
    high (int): Upper bound of the distribution
    
    Returns:
    float: CDF value at x
    """
    if x < low:
        return 0
    elif x >= high:
        return 1
    else:
        # For discrete uniform, the CDF at k is (floor(k) - low + 1) / (high - low + 1)
        return (np.floor(x) - low + 1) / (high - low + 1)

theoretical_cdf_values = [theoretical_discrete_uniform_cdf(x, low, high) for x in x_values]

# Plot both empirical and theoretical CDF
plt.figure(figsize=(10, 6))

# Plot empirical CDF as step function
plt.step(x_values, empirical_cdf_values, 'b-', where='post', label='Empirical CDF', alpha=0.7)

# Plot theoretical CDF as points
plt.step(x_values, theoretical_cdf_values, 'ro-', where='post', label='Theoretical CDF', alpha=0.7, ms=5)

plt.xlabel('x')
plt.ylabel('Cumulative Probability P(X ≤ x)')
plt.title(f'CDF of Discrete Uniform Distribution ({low} to {high})')
plt.xticks(x_values)
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Print some CDF values for verification
print("CDF values at selected points:")
for x in [low-1, low, (low+high)//2, high, high+1]:
    print(f"P(X ≤ {x}) = {theoretical_discrete_uniform_cdf(x, low, high):.4f} (theoretical)")
```

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.
    
    Parameters:
        sample_data (list): list of sample values
        population_mean (float): known population mean (H₀)
        population_std (float): known population standard deviation
        alpha (float): significance level (default = 0.05)
        alternative (str): 'two-sided', 'less', or 'greater'
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / standard_error

    print(f"Sample mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
        z_critical = norm.ppf(1 - alpha / 2)
        reject = abs(z_score) > z_critical
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_score)
        z_critical = norm.ppf(1 - alpha)
        reject = z_score > z_critical
    elif alternative == 'less':
        p_value = norm.cdf(z_score)
        z_critical = norm.ppf(alpha)
        reject = z_score < z_critical
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    print(f"P-value: {p_value:.4f}")
    print(f"Z-critical value: {z_critical:.2f}")
    
    if reject:
        print("✅ Reject the null hypothesis (H₀)")
    else:
        print("❌ Fail to reject the null hypothesis (H₀)")

# 🔢 Example Usage:

# Let's say the population mean (H₀) is 70, and std dev is 10
# We collect a sample and want to test if the mean differs
sample_data = [72, 74, 68, 71, 69, 75, 70, 76, 73, 71]

z_test(sample_data, population_mean=70, population_std=10, alpha=0.05, alternative='two-sided')
```

## Q 16 Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    """
    Calculate and interpret confidence interval for a given dataset.
    
    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default is 0.95)
        
    Returns:
        (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample std dev
    se = std_dev / np.sqrt(n)
    z = norm.ppf((1 + confidence) / 2)
    
    margin_of_error = z * se
    lower = mean - margin_of_error
    upper = mean + margin_of_error
    
    print(f"Sample Mean: {mean:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"📝 Interpretation: We are {int(confidence*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")

    return (lower, upper)

# 🔢 Example usage
sample_data = [52, 55, 60, 58, 62, 59, 57, 61, 64, 63]
confidence_interval(sample_data, confidence=0.95)
```

## Q .1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a Z-test to compare sample mean to population mean.

    Parameters:
        sample (list or np.array): Sample data
        pop_mean (float): Known population mean
        pop_std (float): Known population standard deviation
        alpha (float): Significance level (default 0.05)

    Returns:
        z_stat (float): Calculated Z statistic
        p_value (float): Two-tailed p-value
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    
    # Calculate standard error
    std_error = pop_std / np.sqrt(n)
    
    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / std_error
    
    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    
    # Print results
    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {pop_mean}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")
    
    # Interpretation
    if p_value < alpha:
        print(f"Reject the null hypothesis at alpha = {alpha}.")
        print("There is significant evidence that the sample mean differs from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at alpha = {alpha}.")
        print("There is no significant evidence that the sample mean differs from the population mean.")
    
    return z_stat, p_value

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=102, scale=15, size=30)  # Sample with mean around 102
population_mean = 100
population_std = 15  # Known population std dev

z_test(sample_data, population_mean, population_std)
```

## Q5  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(pop_mean, alt_mean, pop_std, sample_size, alpha=0.05):
    """
    Calculate and visualize Type I and Type II errors.

    Parameters:
        pop_mean (float): Mean under null hypothesis
        alt_mean (float): Mean under alternative hypothesis
        pop_std (float): Known population standard deviation
        sample_size (int): Sample size
        alpha (float): Significance level (default 0.05)

    Returns:
        None
    """
    # Standard error
    std_error = pop_std / np.sqrt(sample_size)
    
    # Critical z-value for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)
    
    # Critical values in original scale
    lower_crit = pop_mean - z_critical * std_error
    upper_crit = pop_mean + z_critical * std_error
    
    # Define x range for plotting
    x_min = pop_mean - 4*std_error
    x_max = alt_mean + 4*std_error
    x = np.linspace(x_min, x_max, 1000)
    
    # Null distribution (H0)
    null_pdf = norm.pdf(x, pop_mean, std_error)
    # Alternative distribution (H1)
    alt_pdf = norm.pdf(x, alt_mean, std_error)
    
    plt.figure(figsize=(12, 6))
    
    # Plot null distribution
    plt.plot(x, null_pdf, label='Null Hypothesis (H0)', color='blue')
    # Plot alternative distribution
    plt.plot(x, alt_pdf, label='Alternative Hypothesis (H1)', color='orange')
    
    # Shade rejection regions (Type I error area under H0)
    plt.fill_between(x, 0, null_pdf, where=(x <= lower_crit), color='red', alpha=0.3, label='Rejection Region (Type I error area)')
    plt.fill_between(x, 0, null_pdf, where=(x >= upper_crit), color='red', alpha=0.3)
    
    # Shade Type II error region (area under H1 in acceptance region)
    plt.fill_between(x, 0, alt_pdf, where=( (x > lower_crit) & (x < upper_crit) ), color='green', alpha=0.3, label='Type II Error Area')
    
    # Mark critical values
    plt.axvline(lower_crit, color='red', linestyle='--', label=f'Lower Critical Value = {lower_crit:.2f}')
    plt.axvline(upper_crit, color='red', linestyle='--', label=f'Upper Critical Value = {upper_crit:.2f}')
    
    plt.title('Type I and Type II Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Calculate Type II error (beta)
    beta = norm.cdf(upper_crit, alt_mean, std_error) - norm.cdf(lower_crit, alt_mean, std_error)
    power = 1 - beta
    
    print(f"Significance Level (alpha, Type I error): {alpha}")
    print(f"Type II error (beta): {beta:.4f}")
    print(f"Power of the test (1 - beta): {power:.4f}")

# Example usage
plot_type1_type2_errors(pop_mean=100, alt_mean=105, pop_std=15, sample_size=30, alpha=0.05)
```

```python
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05):
    # Perform independent t-test
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welch's t-test
    
    print(f"Sample 1 Mean = {np.mean(sample1):.3f}")
    print(f"Sample 2 Mean = {np.mean(sample2):.3f}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")
    
    # Interpretation
    if p_value < alpha:
        print(f"Reject the null hypothesis at alpha = {alpha}.")
        print("There is a significant difference between the two group means.")
    else:
        print(f"Fail to reject the null hypothesis at alpha = {alpha}.")
        print("No significant difference between the two group means.")

# Example usage
np.random.seed(42)
group1 = np.random.normal(loc=100, scale=10, size=30)
group2 = np.random.normal(loc=105, scale=12, size=30)

independent_t_test(group1, group2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test(before, after, alpha=0.05):
    # Perform paired t-test
    t_stat, p_value = ttest_rel(before, after)
    
    print(f"Mean Before = {np.mean(before):.3f}")
    print(f"Mean After = {np.mean(after):.3f}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")
    
    if p_value < alpha:
        print(f"Reject the null hypothesis at alpha = {alpha}.")
        print("There is a significant difference between before and after measurements.")
    else:
        print(f"Fail to reject the null hypothesis at alpha = {alpha}.")
        print("No significant difference between before and after measurements.")
    
    # Visualization
    plt.figure(figsize=(8,6))
    plt.plot(before, label='Before', marker='o')
    plt.plot(after, label='After', marker='o')
    
    # Connect paired points with lines
    for i in range(len(before)):
        plt.plot([i, i], [before[i], after[i]], 'k--', alpha=0.5)
    
    plt.title('Paired Sample Measurements Before and After')
    plt.xlabel('Sample Index')
    plt.ylabel('Measurement')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
np.random.seed(42)
before = np.random.normal(loc=50, scale=5, size=20)
after = before + np.random.normal(loc=2, scale=4, size=20)  # after has a small increase with noise

paired_t_test(before, after)
```

## Q 8 Simulate data and perform both Z-test and T-test, then compare the results using Python

```python
import numpy as np
from scipy.stats import norm, t

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    std_error = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / std_error
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed p-value
    
    print("=== Z-Test ===")
    print(f"Sample mean = {sample_mean:.3f}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")
    if p_value < alpha:
        print(f"Reject null hypothesis at alpha={alpha}: significant difference.")
    else:
        print(f"Fail to reject null hypothesis at alpha={alpha}: no significant difference.")
    print()

def t_test(sample, pop_mean, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    std_error = sample_std / np.sqrt(n)
    t_stat = (sample_mean - pop_mean) / std_error
    df = n - 1
    p_value = 2 * (1 - t.cdf(abs(t_stat), df))  # two-tailed p-value
    
    print("=== T-Test ===")
    print(f"Sample mean = {sample_mean:.3f}")
    print(f"Sample std deviation = {sample_std:.3f}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")
    if p_value < alpha:
        print(f"Reject null hypothesis at alpha={alpha}: significant difference.")
    else:
        print(f"Fail to reject null hypothesis at alpha={alpha}: no significant difference.")
    print()

# Simulate data
np.random.seed(123)
pop_mean = 50
pop_std = 10
sample_size = 30

# Generate sample data with a slight difference from pop_mean
sample_data = np.random.normal(loc=52, scale=pop_std, size=sample_size)

# Perform tests
z_test(sample_data, pop_mean, pop_std)
t_test(sample_data, pop_mean)
```

## Q 9 Write a Python function to calculate the confidence interval for a sample mean and explain its significanceE

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # Standard error of the mean
    
    # t-critical value for two-tailed test
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    
    margin_of_error = t_critical * std_err
    
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    
    print(f"Sample Mean = {mean:.3f}")
    print(f"{int(confidence*100)}% Confidence Interval = ({lower_bound:.3f}, {upper_bound:.3f})")
    
    return lower_bound, upper_bound

# Example usage
import numpy as np
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=40)

confidence_interval(sample_data)
```

## Q 10 Write a Python program to calculate the margin of error for a given confidence level using sample dataE


```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # Standard error
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)  # t critical value
    
    margin_error = t_critical * std_err
    print(f"Sample size = {n}")
    print(f"Standard Error = {std_err:.4f}")
    print(f"{int(confidence*100)}% Confidence Level")
    print(f"Margin of Error = {margin_error:.4f}")
    
    return margin_error

# Example usage
import numpy as np
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=8, size=35)

margin_of_error(sample_data, confidence=0.95)
```

## Q 11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, likelihood, evidence):
    """
    Calculate the posterior probability using Bayes' Theorem.
    
    Parameters:
    prior (float): Prior probability P(H)
    likelihood (float): Likelihood P(E|H)
    evidence (float): Evidence probability P(E)
    
    Returns:
    float: Posterior probability P(H|E)
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example scenario:
# A medical test for a disease:
# - Probability that a person has the disease (prior): 1% = 0.01
# - Probability of a positive test if the person has the disease (likelihood): 99% = 0.99
# - Probability of a positive test in the general population (evidence): 5% = 0.05

prior = 0.01
likelihood = 0.99
evidence = 0.05

posterior = bayes_theorem(prior, likelihood, evidence)

print(f"Posterior probability of having the disease given a positive test: {posterior:.4f}")
```

## Q 12 Perform a Chi-square test for independence between two categorical variables in PythonE

```python
import numpy as np
from scipy.stats import chi2_contingency

# Example: Contingency table of two categorical variables
# Rows: Gender (Male, Female)
# Columns: Preference (Like, Dislike)

# Sample data (observed frequencies)
data = np.array([[30, 10],  # Males who Like and Dislike
                 [20, 40]]) # Females who Like and Dislike

# Perform Chi-square test
chi2, p_value, dof, expected = chi2_contingency(data)

print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)
print("Expected Frequencies:\n", expected)

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print(f"Reject the null hypothesis: Variables are dependent (significant association)")
else:
    print(f"Fail to reject the null hypothesis: Variables are independent (no significant association)")
```

## Q 14  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

observed = np.array([50, 30, 20])

# Define expected proportions (e.g., uniform distribution)
expected_proportions = np.array([1/3, 1/3, 1/3])

# Scale expected frequencies to sum to the observed total
expected = expected_proportions * observed.sum()

# Now perform the test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed data does NOT fit the expected distribution")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution")
```
